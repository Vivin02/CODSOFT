## LABEL ENCODER

```python
#since what the computer understand is 0s and 1s therefore the species are converted into numeric form.
le = LabelEncoder()
iris_data['species']=le.fit_transform(iris_data['species'])
print('Encoded Species: ', iris_data['species'].unique())
iris_data.head()
```

## SPLITTING THE DATA INTO TRAINING AND TESTING MODEL

```python
from sklearn.model_selection import train_test_split
# train - 75
# test - 25
X = iris_data.drop(columns = ['species'])
y = iris_data['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.25)
print(X_train)
print(y_train)
```

## Support Vector machine Algorithm -1

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)

```

```python
pred_1 = model_svc.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, pred_1))
```

## Logistic Regression - 2

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(X_train, y_train)
```

```python
# to print metric to get performance
pred_2 = model_LR.predict(X_test)
print(accuracy_score(y_test, pred_2))
```

## KNeighborsClassifier - 3

```python
# KNN 
from sklearn.neighbors import KNeighborsClassifier
model_KNN= KNeighborsClassifier()
model_KNN.fit(X_train, y_train)
```

```python
# to print metric to get performance
pred_3 = model_KNN.predict(X_test)
print(accuracy_score(y_test, pred_3))
```
